## visualize runs 

```python
%matplotlib inline
from __future__ import division
from pylab import *
import h5py
import matplotlib.pyplot as plt
import math
import numpy as np
import os
import re
```

```python
filename = "sim/bistableMAPKyeast.lm"
fp = h5py.File(filename, "r")
```

```python
ind = 1
cont = True
while (cont):
    try:
        fp["/Simulations/%07d/SpeciesCounts"%ind]
        ind = ind + 1
    except:
        cont = False
print ind
```

```python
replicate = 1
counts=fp["/Simulations/%07d/SpeciesCounts"%replicate]
print "Counts shape and type are:",counts.shape,counts.dtype
print "Initial counts are:",counts[0,:]
print "Final counts are:",counts[-1,:]
```

```python
times=fp["/Simulations/%07d/SpeciesCountTimes"%replicate]
print "Times shape and type are:",times.shape,times.dtype
print "Time begin and end are:",times[0],times[-1]
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(times,counts[:,0])
plt.plot(times,counts[:,1])
plt.plot(times,counts[:,2])
plt.plot(times,counts[:,3])
plt.plot(times,counts[:,4])
plt.plot(times,counts[:,5])
plt.plot(times,counts[:,6])
xlabel('Time (s)'); ylabel('# Molecules')
plt.legend(['RL', 'R', 'L', 'K', 'Kp', 'Kpp', 'P'])
plt.title(''.join(['replicate ', str(replicate)]))
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(times,counts[:,1])
plt.plot(times,counts[:,2])
plt.plot(times,counts[:,3])
plt.plot(times,counts[:,4])
plt.plot(times,counts[:,5])
xlabel('Time (s)'); ylabel('# Molecules')
plt.title('fluctuating variables of interest')
plt.legend(['R', 'L', 'K', 'Kp', 'Kpp'])
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
c = np.asarray(counts)
t = np.asarray(times)
for i in range(replicate, ind):
    counts=fp["/Simulations/%07d/SpeciesCounts"%i]
    times=fp["/Simulations/%07d/SpeciesCountTimes"%i]
    c = (np.asarray(counts) + c)/2.0
    t = (np.asarray(times + t))/2.0
    
plt.plot(t, c)
xlabel('Time (s)'); ylabel('# Molecules')
plt.legend(['RL', 'R', 'L', 'K', 'Kp', 'Kpp', 'P'])
plt.title(''.join(['average of ', str(ind), ' iterations']))

plt.show()
```
