```python
%reset
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as snt
from scipy.optimize import curve_fit
from numpy.linalg import svd
import os
from scipy.optimize import curve_fit
import plotly.plotly as py
import plotly.graph_objs as go
from decimal import Decimal
from matplotlib.colors import LogNorm
```

```python
### SPECIES ###
# S    substrate                        (c1)
# K    kinase                           (c2)
# Kp   phosphorylated kinase            (c3)
# Kpp  doubly phosphorlyated kinase     (c4)

### EQUATIONS FOR CELL SIGNALLING ###
# S + K -> S + Kp  (k1)
# K + K -> Kp + K  (k2)
# Kp+ K -> K + K   (k3)
# Kp+Kp -> Kpp+Kp  (k4)
# Kpp+Kp-> Kp+Kp   (k5)

fig = plt.figure(figsize=(10,8))

conc2 = 100
conc1 = 100
V = 1e-15

#equilibrium concentrations
S = np.mean([conc1, conc2])
Ktot = 1000

n = 100
N = 10000
Kp = np.linspace(0,Ktot,n)
Kpp = np.linspace(0,Ktot,n)

Ks = 0.01
theta = ((1+Ks)**2)/((1-2*Ks)**2) # k4k3/k2k5
print 'theta = ', theta

# kinetic rate constants
adjFact = 1.0/Ktot
k1 = 5e-4*adjFact
k2 = 20*adjFact
k3 = 80*adjFact
k4 = 50*adjFact
k5 = 10*adjFact

#print 'assymetric factor =', (k4*k3)/(k2*k5)

#append to differential space
A = np.matrix(np.asarray([
                -k1, -k2, k3, 0, 0,
                k1, k2, -k3, -k4, k5, 
                0, 0, 0, k4, -k5 ]).reshape([3,5]))
```

```python
#find individual kinase differentials
dk = np.ones(len(Kp)*len(Kpp)).reshape([len(Kpp), len(Kp)])*1000
dkp = np.ones(len(Kp)*len(Kpp)).reshape([len(Kpp), len(Kp)])*1000
dkpp = np.ones(len(Kp)*len(Kpp)).reshape([len(Kpp), len(Kp)])*1000
k = []
for i in range(len(Kp)):
    for j in range(len(Kpp)):
        K = Ktot - Kp[i] - Kpp[j]
        if K >= 0:
            dk[i,j] = np.power(np.matmul(A,np.asarray([S*K,K**2,K*Kp[i],Kp[i]**2,Kp[i]*Kpp[j]]).reshape([5,1])),2)[0]
            dkp[i,j] = np.power(np.matmul(A,np.asarray([S*K,K**2,K*Kp[i],Kp[i]**2,Kp[i]*Kpp[j]]).reshape([5,1])),2)[1]
            dkpp[i,j] = np.power(np.matmul(A,np.asarray([S*K,K**2,K*Kp[i],Kp[i]**2,Kp[i]*Kpp[j]]).reshape([5,1])),2)[2]

#plot
fig = plt.figure(figsize=(18,6))
ax = fig.add_subplot(1,3,1)
plt.imshow(np.log(dk), cmap = 'bwr', origin = 'lower')
plt.xlabel('kpp(index)')
plt.ylabel('kp (index)')
plt.title('dk/dt')
plt.colorbar()
ax = fig.add_subplot(1,3,2)
plt.imshow(np.log(dkp), cmap = 'bwr', origin = 'lower')
plt.xlabel('kpp(index)')
plt.ylabel('kp (index)')
plt.title('dkp/dt')
plt.colorbar()
ax = fig.add_subplot(1,3,3)
plt.imshow(np.log(dkpp), cmap = 'bwr', origin = 'lower')
plt.xlabel('kpp(index)')
plt.ylabel('kp (index)')
plt.title('dkpp/dt')
plt.colorbar()

plt.show()
```

```python
grad = dk + dkp + dkpp 

fig = plt.figure(figsize=(6,6))
plt.imshow(np.log(grad), cmap = 'bwr', origin = 'lower')
plt.xlabel('kpp (index)')
plt.ylabel('kp (index)')
plt.title('kp vs kpp weighted by g|radient magnitude')
plt.colorbar()

plt.show()
```

```python
#nullspace function courtesy of scipy
def nullspace(A, atol=1e-13, rtol=0):
    A = np.atleast_2d(A)
    u, s, vh = svd(A)
    tol = max(atol, rtol * s[0])
    nnz = (s >= tol).sum()
    ns = vh[nnz:].conj().T
    return ns
```

```python
# run some ODEs for steady state solutions just found

T = 10

def rates(y, t, k1, k2, k3, k4, k5, S):
    return np.asarray([-k1*S*y[0] - k2*y[0]**2 + k3*y[1]*y[0],
                      k1*S*y[0] + k2*y[0]**2 - k3*y[1]*y[0] - k4*y[1]**2 + k5*y[2]*y[1],
                      k4*y[1]**2 -  k5*y[2]*y[1]])

odeTrials = {}
acc = []
newinds = []
ind = 0
t = np.linspace(0,T,N)
params = tuple([k1,k2,k3,k4,k5,S])
for i in range(len(inds)):
    y0 = [Ktot - Kp[inds[i][0]] - Kpp[inds[i][1]], Kp[inds[i][0]], Kpp[inds[i][1]]]
    ode = snt.odeint(rates, y0,t, args = tuple(params))
    odeTrials[ind] = []
    odeTrials[ind] = ode
    acc.append(params)
    ind = ind+ 1
```

```python
def nullparam(S,K,kp, kpp):
    mat = np.matrix([
            [-k1*S*K, -k2*K**2, k3*K*kp, 0, 0], 
            [k1*S*K, k2*K**2, -k3*K*kp, -k4*kp**2, k5*kp*kpp], 
            [0, 0, 0, k4*kp**2, -k5*kp*kpp]
              ])
    nullmat = nullspace(mat)
    for i in range(nullmat.shape[1]): #IDK what the range is for so figure that out nicole
        if len(nullmat) == sum(nullmat[:,i] > 0):
            return nullmat[:,i]
    return np.zeros(len(nullmat))

odeTrials = {}
acc = []
newinds = []
ind = 0
t = np.linspace(0,T,N)
for i in range(len(inds)):
    temp = nullparam(S, Ktot - Kp[inds[i][0]] - Kpp[inds[i][1]], Kp[inds[i][0]], Kpp[inds[i][1]])
    y0 = [float(Ktot)/2,3*float(Ktot)/4,float(Ktot)/4]
    if sum(temp > 0):
        params = []
        for i in range(len(temp)):
            params.append(temp.tolist()[i][0])
        params.append(S)
        ode = snt.odeint(rates, y0,t, args = tuple(params))
        odeTrials[ind] = []
        odeTrials[ind] = ode
        acc.append(params)
        ind = ind+ 1

print ind
```

```python
## attempt at new model using michalelis menten kinetics

## species
# S (signal ~ alpha factor)
# L (receptor ligand)
# K (kinase in MAPK pathway)
# Kp (singly phosphorylated kinase in MAPK)
# Kpp (doubly phosphorylated kinase in MAPK) 

# st Ktot = K + Kp + Kpp 

## kinetic rate laws
# S + L <-> SL             (k1, k_1) 
# SL + K -> SL + Kp        ~similar kinetics to k3
# K <-> Kp                 (k3, k_3)
# Kp <-> Kpp               (k4, k_4)

## differential equations
# dL/dt = -k1*S*L + k_1*V0*S*L
# dK/dt = (V3*(Ktot-K-Kpp))/(K3+(Ktot-K-Kpp)) - (V_3*V0*S*L*K)/(V_3+K)
# dKpp/dt = (V4*(Ktot-K-Kpp))/(K4+(Ktot-K-Kpp)) - (V_4*Kpp)/(K_4+Kpp)


#equilibrium concentrations
S = 1
L = 1
Ktot = 1000

n = 100
N = 10000
K = np.linspace(0,Ktot,n)
Kpp = np.linspace(0,Ktot,n)

# kinetic rate constants
adjFact = 1.0/Ktot
k1 = 1
k_1 = 1
V0 = k1/k_1
V3 = 5
K3 = 5
V_3 = 2.0/V3
K_3 = 2.0/K3
V4 = 10
K4 = 10
V_4 = 2.0/V4
K_4 = 2.0/K4


#find individual kinase differentials
dk = np.ones(len(K)*len(Kpp)).reshape([len(Kpp), len(K)])*1000
dkpp = np.ones(len(K)*len(Kpp)).reshape([len(Kpp), len(K)])*1000
k = []
for i in range(len(K)):
    for j in range(len(Kpp)):
        Kp = Ktot - K[i] - Kpp[j]
        if Kp >= 0:
            dk[i,j] = (V3*Kp)/(K3+Kp) - (V3*V0*S*L*K[i])/(V3+K[i])
            dkpp[i,j] = (V4*Kp)/(K4+Kp) - (V_4*Kpp[j])/(K_4+Kpp[j])

#plot
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(1,2,1)
plt.imshow(np.log(dk), cmap = 'bwr', origin = 'lower')
plt.xlabel('kpp(index)')
plt.ylabel('kp (index)')
plt.title('dk/dt')
plt.colorbar()
ax = fig.add_subplot(1,2,2)
plt.imshow(np.log(dkpp), cmap = 'bwr', origin = 'lower')
plt.xlabel('kpp(index)')
plt.ylabel('kp (index)')
plt.title('dkpp/dt')
plt.colorbar()

plt.show()

```
