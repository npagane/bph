```python
# read in libraries necessary to produce graphics
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as snt
import random
```

```python
fig, ax = plt.subplots(1, 2)
# define intital values and parameters
y0 = [5e-7, 0] #volume
k1 = 5e5
k2 = .2
ti = 0
tf = 10
Vtot = 1e-15
nA = 6.022e23

t = np.linspace(ti, tf, 100)
# define rates equations
def rates(y, t):
    return np.asarray([-k1*y[0]*y[0] + 2*k2*y[1],
               k1*y[0]*y[0] - 2*k2*y[1]])
    
res = snt.odeint(rates, y0,t)

fig.set_size_inches(12,6)

ax[0].plot(t, res[:,0],t, res[:,1]) # page 205

ax[0].set_xlabel('t (s)')
ax[0].set_ylabel('concentration (M)');

ax[0].legend(['[P]', '[P2]'])

#stochastic solution
N = {}
N['Pre'] = np.asarray([2,0,0,1]).reshape([2,2])
N['Post'] = np.asarray([0,1,2,0]).reshape([2,2])
N['M'] = [301,0]
N['c'] = [2*k1/(nA*Vtot),k2]

def rates(x, t):
    return np.asarray([N['c'][0]*x[0]*(x[0]-1)/2,
                       N['c'][1]*x[1]]).reshape([2,1])

def gillespie(N,n):
    tt = 0
    x = N['M']
    S = np.transpose(N['Post']-N['Pre'])
    u = S.shape[0]
    v = S.shape[1]
    tvec = np.zeros(n)
    xmat = np.matrix(np.zeros([n+1,u]))
    xmat[0,:] = x
    for i in range(0,n):
        h = rates(x,tt)
        h0 = np.sum(h)
        #a = np.asarray(N['c'])*h;
        #a0 = np.sum(a);
        #rand = random.random()
        #tau = (1/a0)*np.log(1/rand);
        #tt = tt + tau
        tt = tt + np.random.exponential(1/h0,1)
        #j = np.random.choice(v, 1, (h/np.sum(h)).tolist())
        #x = x + np.transpose(S[:,j])
        #tvec[i] = tt
        #xmat[i+1,:] = x
        #x = x[0].tolist()
        cdf = np.cumsum(h)/h0;
        rand = np.random.random()
        for j in range(len(cdf)):
            if rand < cdf[j]:
                mu = j;
                break
        x = x + np.transpose(S[:,mu])
        tvec[i] = tt
        xmat[i+1,:] = x
    return np.asarray([tvec, xmat])

n = 500
out = gillespie(N, n)

ax[1].step(out[0], out[1][:,0][0:n])
ax[1].step(out[0], out[1][:,1][0:n])

#show plot
fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
#stochastic solution
N = {}
N['Pre'] = np.asarray([2,0,0,1]).reshape([2,2])
N['Post'] = np.asarray([0,1,2,0]).reshape([2,2])
N['M'] = [301, 0]
N['c'] = [1.66e-3,.2]

def rates(x):
    return np.asarray([N['c'][0]*x[0]*(x[0]-1)/2,
                       N['c'][1]*x[1]]).reshape([2,1])

def gillespie(N,n):
    tt = 0
    x = N['M']
    S = np.transpose(N['Post']-N['Pre'])
    u = S.shape[0]
    v = S.shape[1]
    tvec = np.zeros(n)
    xmat = np.matrix(np.zeros([n+1,u]))
    xmat[0,:] = x
    for i in range(0,n):
        h = rates(x)
        h0 = np.sum(h)
        tt = tt + np.random.exponential(1/h0,1)
        #j = np.random.choice(v, 1, (h/h0).tolist())
        #x = x + np.transpose(S[:,j])
        #tvec[i] = tt
        #xmat[i+1,:] = x
        #x = x[0].tolist()
        cdf = np.cumsum(h)/h0;
        rand = np.random.random()
        for j in range(len(cdf)):
            if rand < cdf[j]:
                mu = j;
                break
        x = x + np.transpose(S[:,mu])
        tvec[i] = tt
        xmat[i+1,:] = x
    return np.asarray([tvec, xmat])

ax = fig.gca()
n = 500
trials = {}
trialN = 10000
for i in range(trialN):
    out = gillespie(N, n)
    trials[i] = np.mean(out[1][:,0][0:n])
    ax.step(out[0], out[1][:,0][0:n])
    
ax.set_xlim([0,15])
ax.set_ylim([0, 300])
fig.tight_layout()

# histogram
fig = plt.figure(figsize=(6,6))
ax = fig.gca()
mean = []
for i in range(trialN):
    mean.append(trials[i])
    
plt.hist(mean, bins = 100, normed = True)

#show plot
plt.show()
```

```python
# define intital values and parameters
y0 = [5e-7, 2e-7,0,0] #concentration
k1 = 1e6
k2 = 1e-4
k3 = .1
ti = 0
tf = 50
Vtot = 1e-15

t = np.linspace(ti, tf, 100)
# define rates equations
def rates(y, t):
    return np.asarray([k2*y[2] -k1*y[0]*y[1],
                       (k2+k3)*y[2] - k1*y[0]*y[1],
                       k1*y[0]*y[1] - (k2+k3)*y[2],
                       k3*y[2]])
    
res = snt.odeint(rates, y0,t)

fig = plt.figure(figsize=(6,6))
ax = fig.gca()

ax.plot(t, res[:,0],t, res[:,1], t, res[:,2], 
       t, res[:,3]) # page 209

ax.set_xlabel('t (s)')
ax.set_ylabel('concentration (M)');

ax.legend(['[S]', '[E]', '[SE]', 'P'])
plt.show()
```
