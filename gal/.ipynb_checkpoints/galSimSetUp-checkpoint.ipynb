## reactions and parameters
Below is a simple model of the GAL switch in yeast WITHOUT nucleosome modification. It is not spatially resolved and probably won't need to be...?<br>

Additional features to verify with more literature consensus:<br>
Gal3p sequestering of Gal80p in cytoplasm to inhibit dimerization (seen in PMID:23858060). Currently, only having activated gal3p inhibit the dimerization reaction. Also not allowing gal1p this same ability since no evidence for this thus far.<br>

```python
# rate constants
# these rates were taken from (Bianchi 2018)
# the orignal rates are stochastic with units of 1/min or 1/(mol*min) with cell volume taken into account
vol = 35.7e-15 # L
avo = 6.022e23
norm1 = 1.0/60 # convert to seconds
norm2 = 1.0/(vol*60*avo) # convets to 1/(mol*sec)

# listed below as seen in Bianci et al but scaled by norm factor to make deterministic rates
kgalEg2b = 0.00031353*norm2 # extracellular galactose binds to gal2p
kgalEg2u = 2392.5*norm1 # extracellular galactose unbinds from gal2p
kgalIg2b = 0.00031353*norm2 # intracelllular galactose binds to gal2p
kgalIg2u = 2392.5*norm1 # intracellular galactose unbinds from gal2p
kgalIn = 4350*norm1 # import of galactose bound to g2p inside cell
kgalOut = 4350*norm1 # export of galactose bound to g2p outside cell
kgalIg3b = 7.45e-7*norm2 # binding of galactose to gal3p
kgalIg3u = 890*norm1 # unbinding of galactose from gal3p
kg2transl = 13.4779*norm1 # translation of gal2 RNA 
kg2deg = 0.003851*norm1 # degredation of gal2p
kgal2g4b = 0.1*norm2 # binding of gal4 dimer to gal2
kgal2g4u = 10.101*norm1 # unbinding of gal4 dimer from gal2
kg2transc = 2.54154*norm1 # transcription of activated gal2g4
kr2deg =  0.077016*norm1 # degredation of gal2 RNA
kg1transl = 1.92541*norm1 # translation of gal1 RNA
kgalg1b = 0.000402426*norm2 # binding of galactose to gal1p
kgalg1u = 1842.5*norm1 # unbinding of galactose from gal1p
kg1pdeg = 0.003851*norm1 # degredation of gal1p
kgal1g4b = 0.1*norm2 # binding of gal1 to gal4p
kgal1g4u = 0.384615*norm1 # unbinding of gal4p from gal1
kg1transc = 0.7379*norm1 # transcription of activated gal1g4 (USING THIS RATE FOR GAL10 AND GAL7 TOO)
kr1deg = 0.02236*norm1 # degredation of gal1 RNA
kg3transl = 55.4518*norm1 # translation of gal3 RNA
kgalg3b = kgalIg3b*norm2 # binding of galactose with gal3p
kgalg3u = kgalIg3u*norm1 # unbinding of galactose from gal3p
kg3pdeg = 0.01155*norm1 # degredation of gal3p
kgal3g4b = 0.1*norm2 # binding of gal3 to gal4p
kgal3g4u = 4.03226*norm1 # unbinding of gal4p from gal3
kg3transc = 0.426572*norm1 # transcription of activated gal3g4
kr3deg = 0.02666*norm1 # degredation of gal3 RNA
kg4transl = 10.7091*norm1 # translaiton of gal4 RNA
kg4b = 100*norm2 # dimerization of gal4p
kg4u = 0.001*norm1 # monomerization of gal4p
kg4pdeg = 0.0069315*norm1 # degredation of gal4p
kg4transc = 0.009902*norm1 # transcription of gal4 (DEPENDENT ON MEDIA)
kr4deg = 0.0247552*norm1 # degredation of gal4 RNA
kg80transl = 3.67368*norm1 # translation of gal80 RNA
kg80b = 100*norm2 # dimerization of gal80p
kg80u = 0.001*norm1 # monomerization of gal80p
kg80pdeg = 0.0069315*norm1 # degredation of gal80p
kgal80g4b = 0#0.1*norm2 # binding of gal80 to gal4p
kgal80g4u = 0#4.03226*norm1 # unbinding of gal4p from gal80
kg80transc = 0.6065*norm1 # transcription of activated gal80g4
kr80deg = 0.028881*norm1 # degredation of gal80 RNA
kg7transl = kg1transl # translation of gal7 RNA *GUESS BUT SAME MAGNITUDE*
kgalPg7b = 1e-3*norm2 # binding of gal7p to galP *GUESS BUT SAME MAGNITUDE*
kgalPg7u = 1e3*norm1 # unbinding of gal7p from galP *GUESS BUT SAME MAGNITUDE*
kg7pdeg = 1e-2*norm1 # degredation of gal7p *GUESS BUT SAME MAGNITUDE*
kr7deg = 2e-2*norm1 # degredation of r7 *GUESS BUT SAME MAGNITUDE*
kg10transl = kg1transl # translation of gal10 RNA *GUESS BUT SAME MAGNITUDE*
kg10pdeg = kg7pdeg # degredation of gal10p *GUESS BUT SAME MAGNITUDE*
kr10deg = kr7deg # degredation of gal10 RNA *GUESS BUT SAME MAGNITUDE*
kg4deg = 0.0069315*norm1 # degredation of gal4 dimer
kg80deg = 0.0069315*norm1 # degredation of gal80 dimer
kgal1g80b = 0.1*norm2 # binding of gal1g4 to gal80p
kgal1g80u = 0.085317*norm1 # unbinding of gal80p from gal1g4
kgal2g80b = 0.1*norm2 # binding of gal2g4 to gal80p
kgal2g80u = 0.134989*norm1 # unbinding of gal80p from gal2g4
kgal3g80b = 0.1*norm2 # binding of gal3g4 to gal80p
kgal3g80u = 0.530504*norm1 # unbinding of gal80p from gal3g4
kgal80g80b = 0.1*norm2 # binding of gal80g4 to gal80p
kgal80g80u = 0.530504*norm1 # unbinding of gal80p from gal80g4
kgalkin = 1*norm1#3350*norm1 # phosphoyrlation of galactose to galP by gal1p *GUESS*
kg80g1b = 0#1e-2*norm2 # binding of activated gal1p to gal80p dimer (indiscriminate of gene) *GUESS*
kg80g1u = 0#4e-2*norm1 # unbinding of activated gal1p from gal80p dimer (indiscriminate of gene) *GUESS*
kg80g3b = 0.1*norm2 # binding of activated gal3p to gal80p dimer (indiscriminate of gene) *GUESS*
kg80g3u = 0*norm1 # unbinding of activated gal3p from gal80p dimer (indiscriminate of gene) *GUESS*
kg80g3diss = 0*norm1 # dissciation of gal80p-gal3p from a gene activated by gal4p *BIGGEST GUESS*
ktrans = 0#1*norm1 # gal7p-galP + UDPGlucose transfer reaction to make UDPGal and gluP *GUESS*
kg7deg = 1e-2*norm1 # degredation of gal7p activated by galP *GUESS*

# applicable initial concentrations also taken from bianchi paper
# conversion factor from molecules to molarity

conv = 1.0/(avo*vol)
# initials
galE = 6.0
galI = 0
galEg2 = 0
galIg2 = 0
gal2p = 1156.91017705*conv
gal2 = 0
r2 = 1*conv
gal1p = 132.318563461*conv
gal1 = 0
r1 = 1*conv
gal3p = 4341.70321121*conv
gal3 = 0
r3 = 1*conv
gal4p = 1*conv
r4 = 1*conv
gal80p = 1*conv
gal80 = 0
r80 = 1.18715948592*conv
gal7p = 0
r7 = 1*conv
gal10p = 0
r10 = 1*conv
g4 = 308.921734356*conv
g80 = 157.246650776*conv
g1 = 0
g3 = 0
g80g3 = 0
gal1g4 = 0
gal1g4g80 = 1*conv
gal1g4g80g1 = 0
gal1g4g80g3 = 0
gal2g4 = 0
gal2g4g80 = 1*conv
gal2g4g80g1 = 0 
gal2g4g80g3 = 0
gal3g4 = 0
gal3g4g80 = 1*conv
gal3g4g80g1 = 0
gal3g4g80g3 = 0
gal80g4 = 0
gal80g4g80 = 1*conv
gal80g4g80g1 = 0
gal80g4g80g3 = 0
galP = 0
gluP = 0
g7 = 0
udpGlu = 0
udpGal = 0

# define reaction rates function
def rates(spec, t):
    # assign spec indexing to metabolite names
    [galE, galI, galEg2, galIg2, gal2p, gal2, r2, gal1p, gal1, r1, gal3p, gal3, r3, gal4p, r4, gal80p, gal80, r80, 
     gal7p, r7, gal10p, r10, g4, g80, g1, g3, g80g3, gal1g4, gal1g4g80, gal1g4g80g1, gal1g4g80g3,
     gal2g4, gal2g4g80, gal2g4g80g1, gal2g4g80g3, gal3g4, gal3g4g80, gal3g4g80g1, gal3g4g80g3, 
     gal80g4, gal80g4g80, gal80g4g80g1, gal80g4g80g3, galP, gluP, g7, udpGlu, udpGal] = spec
    # reactions
    return np.asarray([
        0, #-kgalEg2b*galE*gal2p + kgalEg2u*galEg2, # galE
        -kgalIg2b*galI*gal2p + kgalIg2u*galIg2 - kgalg1b*galI*gal1p + kgalg1u*g1 - kgalIg3b*galI*gal3p + kgalIg3u*g3, # galI
        kgalEg2b*galE*gal2p - kgalEg2u*galEg2 - kgalIn*galEg2 + kgalOut*galIg2, # galEg2
        kgalIg2b*galI*gal2p - kgalIg2u*galIg2 - kgalOut*galIg2 + kgalIn*galEg2, # galIg2
        kg2transl*r2 - kgalEg2b*galE*gal2p - kgalIg2b*galI*gal2p + kgalEg2u*galEg2 + kgalIg2u*galIg2 - kg2deg*gal2p, # gal2p 
        -kgal2g4b*gal2*g4 + kgal2g4u*gal2g4, # gal2
        kg2transc*gal2g4 - kr2deg*r2, # r2 
        kg1transl*r1 - kgalg1b*galI*gal1p + kgalg1u*g1 - kg1pdeg*gal1p, # gal1p
        -kgal1g4b*gal1*g4 + kgal1g4u*gal1g4, # gal1 
        kg1transc*gal1g4 - kr1deg*r1, # r1
        kg3transl*r3 - kgalg3b*galI*gal3p + kgalg3u*g3 - kg3pdeg*gal3p, # gal3p
        -kgal3g4b*gal3*g4 + kgal3g4u*gal3g4, # gal3
        kg3transc*gal3g4 - kr3deg*r3, # r3
        kg4transl*r4 - kg4b*gal4p*gal4p + kg4u*g4 - kg4pdeg*gal4p, # gal4p  
        kg4transc - kr4deg*r4, # r4 
        kg80transl*r80 - kg80b*gal80p*gal80p + kg80u*g80 - kg80pdeg*gal80p, # gal80p 
        -kgal80g4b*gal80*g4 + kgal80g4u*gal80g4, # gal80 
        kg80transc*gal80g4 - kr80deg*r80, # r80
        kg7transl*r7 - kgalPg7b*galP*gal7p + kgalPg7u*g7 - kg7pdeg*gal7p, # gal7p
        kg1transc*gal1g4 - kr7deg*r7, # r7
        kg10transl*r10 - kg10pdeg*gal10p, # gal10p 
        kg1transc*gal1g4 - kr10deg*r10, # r10 
        kg4b*gal4p*gal4p - kg4u*g4 - kgal1g4b*gal1*g4 + kgal1g4u*gal1g4 - kgal2g4b*gal2*g4 + kgal2g4u*gal2g4
            - kgal3g4b*gal3*g4 + kgal3g4u*gal3g4 - kgal80g4b*gal80*g4 + kgal80g4u*gal80g4 - kg4deg*g4, # g4 
        kg80b*gal80p*gal80p - kg80u*g80 - kgal1g80b*gal1g4*g80 + kgal1g80u*gal1g4g80 - kgal2g80b*gal2g4*g80 
            + kgal2g80u*gal2g4g80 - kgal3g80b*gal3g4*g80 + kgal3g80u*gal3g4g80 - kgal80g80b*gal80g4*g80 
            + kgal80g80u*gal80g4g80 - kg80deg*g4 + kg80g3u*g80g3, # - kg80g3b*g3*g80, # g80
        kgalg1b*galI*gal1p - kgalg1u*g1 - kgalkin*g1 - kg80g1b*gal1g4g80*g1 + kg80g1u*gal1g4g80g1
            - kg80g1b*gal2g4g80*g1 + kg80g1u*gal2g4g80g1 - kg80g1b*gal3g4g80*g1 + kg80g1u*gal3g4g80g1
            - kg80g1b*gal80g4g80*g1 + kg80g1u*gal80g4g80g1, # g1
        kgalg3b*galI*gal3p - kgalg3u*g3 + kg80g3u*g80g3 - kg80g3b*gal1g4g80*g3 
            + kg80g3u*gal1g4g80g3 - kg80g3b*gal2g4g80*g3 + kg80g3u*gal2g4g80g3 - kg80g3b*gal3g4g80*g3
            + kg80g3u*gal3g4g80g3 - kg80g3b*gal80g4g80*g3 + kg80g3u*gal80g4g80g3, # - kg80g3b*g3*g80 # g3
        - kg80g3u*g80g3 + kg80g3diss*gal1g4g80g1 + kg80g3diss*gal1g4g80g3 + kg80g3diss*gal2g4g80g1
            + kg80g3diss*gal2g4g80g3 + kg80g3diss*gal3g4g80g1 + kg80g3diss*gal3g4g80g3 + kg80g3diss*gal80g4g80g1
            + kg80g3diss*gal80g4g80g3, #+ kg80g3b*g3*g80 , # g80g3
        kgal1g4b*gal1*g4 - kgal1g4u*gal1g4 - kgal1g80b*gal1g4*g80 + kgal1g80u*gal1g4g80 + kg80g3diss*gal1g4g80g1
            + kg80g3diss*gal1g4g80g3, # gal1g4 
        kgal1g80b*gal1g4*g80 - kgal1g80u*gal1g4g80 - kg80g1b*gal1g4g80*g1 + kg80g1u*gal1g4g80g1
            - kg80g3b*gal1g4g80*g3 + kg80g3u*gal1g4g80g3, # gal1g4g80 
        kg80g1b*gal1g4g80*g1 - kg80g1u*gal1g4g80g1 - kg80g3diss*gal1g4g80g1, # gal1g4g80g1
        kg80g3b*gal1g4g80*g3 - kg80g1u*gal1g4g80g3 - kg80g3diss*gal1g4g80g3, # gal1g4g80g3
        kgal2g4b*gal2*g4 - kgal2g4u*gal2g4 - kgal2g80b*gal2g4*g80 + kgal2g80u*gal2g4g80 + kg80g3diss*gal2g4g80g1
            + kg80g3diss*gal2g4g80g3, # gal2g4 
        kgal2g80b*gal2g4*g80 - kgal2g80u*gal2g4g80 - kg80g1b*gal2g4g80*g1 + kg80g1u*gal2g4g80g1
            - kg80g3b*gal2g4g80*g3 + kg80g3u*gal2g4g80g3, # gal2g4g80 
        kg80g1b*gal2g4g80*g1 - kg80g1u*gal2g4g80g1 - kg80g3diss*gal2g4g80g1, # gal2g4g80g1
        kg80g3b*gal2g4g80*g3 - kg80g1u*gal2g4g80g3 - kg80g3diss*gal2g4g80g3, # gal2g4g80g3
        kgal3g4b*gal3*g4 - kgal3g4u*gal3g4 - kgal3g80b*gal3g4*g80 + kgal3g80u*gal3g4g80 + kg80g3diss*gal3g4g80g1
            + kg80g3diss*gal3g4g80g3, # gal3g4 
        kgal3g80b*gal3g4*g80 - kgal3g80u*gal3g4g80 - kg80g1b*gal3g4g80*g1 + kg80g1u*gal3g4g80g1
            - kg80g3b*gal3g4g80*g3 + kg80g3u*gal3g4g80g3, # gal3g4g80 
        kg80g1b*gal3g4g80*g1 - kg80g1u*gal3g4g80g1 - kg80g3diss*gal3g4g80g1, # gal3g4g80g1
        kg80g3b*gal3g4g80*g3 - kg80g1u*gal3g4g80g3 - kg80g3diss*gal3g4g80g3, # gal3g4g80g3
        kgal80g4b*gal80*g4 - kgal80g4u*gal80g4 - kgal80g80b*gal80g4*g80 + kgal80g80u*gal80g4g80 + kg80g3diss*gal80g4g80g1
            + kg80g3diss*gal80g4g80g3, # gal80g4 
        kgal80g80b*gal80g4*g80 - kgal80g80u*gal80g4g80 - kg80g1b*gal80g4g80*g1 + kg80g1u*gal80g4g80g1
            - kg80g3b*gal80g4g80*g3 + kg80g3u*gal80g4g80g3, # gal80g4g80 
        kg80g1b*gal80g4g80*g1 - kg80g1u*gal80g4g80g1 - kg80g3diss*gal80g4g80g1, # gal80g4g80g1
        kg80g3b*gal80g4g80*g3 - kg80g1u*gal80g4g80g3 - kg80g3diss*gal80g4g80g3, # gal80g4g80g3
        kgalkin*g1 - kgalPg7b*galP*gal7p + kgalPg7u*g7, # galP
        ktrans*g7*udpGlu, # gluP
        kgalPg7b*galP*gal7p - kgalPg7u*g7 - ktrans*g7*udpGlu - kg7deg*g7, # g7
        gal10p*udpGal - g7*udpGlu, # udpGlu
        ktrans*g7*udpGlu - gal10p*udpGal, # udpGal
    ])
```

```python
# solve the DE numerically
t = np.linspace(0,1*60*60, 100000)
spec0 = (galE, galI, galEg2, galIg2, gal2p, gal2, r2, gal1p, gal1, r1, gal3p, gal3, r3, gal4p, r4, gal80p, gal80, r80, 
     gal7p, r7, gal10p, r10, g4, g80, g1, g3, g80g3, gal1g4, gal1g4g80, gal1g4g80g1, gal1g4g80g3,
     gal2g4, gal2g4g80, gal2g4g80g1, gal2g4g80g3, gal3g4, gal3g4g80, gal3g4g80g1, gal3g4g80g3, 
     gal80g4, gal80g4g80, gal80g4g80g1, gal80g4g80g3, galP, gluP, g7, udpGlu, udpGal)
ode = snt.odeint(rates, spec0, t)
```

```python
fig = plt.figure(figsize=(40,30))

# plot the numerical solution
specs = ['galE', 'galI', 'galEg2', 'galIg2', 'gal2p', 'gal2', 'r2', 'gal1p', 'gal1', 'r1', 'gal3p', 'gal3', 'r3', 'gal4p', 
         'r4', 'gal80p', 'gal80', 'r80', 'gal7p', 'r7', 'gal10p', 'r10', 'g4', 'g80', 'g1', 'g3', 'g80g3', 'gal1g4', 
         'gal1g4g80', 'gal1g4g80g1', 'gal1g4g80g3', 'gal2g4', 'gal2g4g80', 'gal2g4g80g1', 'gal2g4g80g3', 'gal3g4', 
         'gal3g4g80', 'gal3g4g80g1', 'gal3g4g80g3', 'gal80g4', 'gal80g4g80', 'gal80g4g80g1', 'gal80g4g80g3', 'galP', 
         'gluP', 'g7', 'udpGlu', 'udpGal']
#plot ligand 
for i in range(np.shape(ode)[1]):
    ax = fig.add_subplot(int(np.shape(ode)[1]/4),4, i+1)
    ax.plot(t, ode[:, i], linewidth = 4)
    ax.legend([specs[i]], fontsize = 'large')

plt.show()
```
