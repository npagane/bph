## determining energy of RNAP passage near nucleosome

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 2
end = r*2*np.pi

def loop(x):
    return -r*np.sin(x/r) + 1.0/x - 1.55*r

x = np.linspace(1e-3, end)
plt.plot(x, loop(x))

plt.plot(x, np.ones(len(x))*r)
plt.ylim([-end/2.0, end/2.0])

a = 1; b = 2.55*r; c = -1
start = (-b + np.sqrt(b**2 - 4*a*c))/2*a
plt.plot(start, loop(start), 'o')

stop = np.argmax(-r*np.sin(x/r) - 1.55*r)
stop = x[stop]
plt.plot(stop, loop(stop), 'o')

print 'dist ', r - loop(stop)

plt.show()
```

```python
from scipy.integrate import quad

def dloop(x):
    return -np.cos(x/r) - 1.0/x**2

def integrand(x):
    return np.sqrt(1 + dloop(x)**2)

I = quad(integrand, start, end)
L = I[0] + end
base = end/L
print L, base
```

```python
bp = 90

def rvec(s):
    if s <= base*L:
        x = end-end*s/(base*L); y = 0
    else:
        x = end - end*s/((1-base)*L); y = loop(x)
        if y > r: y = r
    return np.sqrt(x**2 + y**2)

x = np.linspace(0, L, bp)
ar = []
for i in x:
    ar.append(rvec(i))
    
print base*L
    
plt.plot(x, ar)
plt.show()
```

```python
# second-order central differencing 
def cd2(s):
    h = 1e-8 # best h for central differencing
    return (rvec(s + h) - 2*rvec(s) + rvec(s - h))/h**2


l = np.linspace(0, L, bp)
dr2 = []
for i in l:
    dr2.append(cd2(i))
    
plt.plot(dr2)
plt.show()
```

```python
N = 1e3; a = 0.0; b = L
h = (b-a)/N
# determine odd sum 
k = np.linspace(1, N-1, int(N/2))
oddSum = 0
for i in k:
    oddSum += cd2(a + i*h)
# determine even sum
k = np.linspace(2, N-2, int(N/2) - 1)
evenSum = 0
for i in k:
    evenSum += cd2(a + i*h)
# find final area
integ = 1.0/(3)*h *(cd2(a) + cd2(b) + 4*oddSum + 2*evenSum)
```

```python
kB = 1.38e-23; T = 295
E = 2e9 # youngs modulus
I = (np.pi*(1e-9)**4)/4 # moment of inertia of rod
Bs = E*I
P = Bs/(kB*T)
print P*1e9/3
```

```python
Etor = 0.5*kB*T*P*integ
print Etor
```

```python
kspring = 2*Etor/(1.0/bp*3*1e-9)**2
```

```python
def Etrans(spc):
    return 0.5*kspring*(1.0/spc*3*1e-9)**2

x = np.linspace(1e-4, 200)
plt.plot(x, Etrans(x))
plt.plot(x, kB*T*np.ones(len(x)))
plt.plot(x, Etor*np.ones(len(x)))
plt.ylim([0, 5*kB*T])

plt.xlabel('spacing (bp)')
plt.ylabel('E (J)')

plt.legend(['spring', 'kT', 'Etor'])

plt.show()
```
